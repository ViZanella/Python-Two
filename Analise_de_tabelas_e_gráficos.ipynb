```python
def GraficoLinha(style='ggplot'):
  #estilo de cor
  plt.style.use(style) 

  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 16})

  # Criar um gráfico de linha
  plt.figure(figsize=(16, 12))

  #descompactar lista
  produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o') 
  produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
  produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')


  plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
  plt.title('Vendas de Produtos (2023)')
  plt.xlabel('Mês')
  plt.ylabel('Vendas')

  plt.savefig('GraficoLinha.svg')
  # Mostrar o gráfico

  plt.show()
```

```python
df = pd.read_csv('Idade_Carro.csv')
df
```

```python
# Criar o gráfico de dispersão
plt.scatter(df['Idade'], df['Preço'])

plt.title('Preço do Carro x Idade')
plt.xlabel('Idade do Carro (Anos)')
plt.ylabel('Preço (R$)')

# Mostrar o gráfico
plt.show()
```

# Explique como esse gráfico:

O grafico mostra um correlação entre as duas variaveis com correlação decendente, mostrando como foi a alteração do preço de um carro que diminuiu levando em conta seu ano.

# Por que é interessante fazer um gráfico de dispersão nesse caso?
R:Demonstrar há alteração no preço de veiculos(linha Y)conforme o ano(linha x) que eles vão saindo de linha e novos carros mais modernos são lançados, montrando  que quanto mais velho o carro mais o valor dele cai.

# Qual é a diferença entre o gráfico de linha na interpretação?
R:O gráfico de linha tem seus pontos conectados , enquanto o gráfico de dispersão mostra os pontos de forma indivicual.No gráfico de linha é mostrada a mudança de uma variável ao longo do tempo ou de uma sequência,ja no gráfico de dispersão mostra a relação entre duas variáveis independentes, ou seja a forma como eles vão tratar os dados baseado no eixo horizontal.

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.normal(loc=50, scale=15, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: No modelo distribuição, ela mostra a distribuição de dados da variavel,a forma como estão distruibuidos em diferentes categorias ou intervalos.
No modelo normal é medido o desvio padrão determinando a sua localização e dispersão.

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: O grafico indica a discrepancia entre os dados podendo ter um valor minimo de aproximadamente -10 e máximo de aproximadamente 90, e o primeiro quartil está localizado entre 24 e 42 o segundo quartil/mediana é um valor aproximado de 43/44 e o terceiro quartil sendo de aproximadamente entre 44 e 56, onde os quartis representão respectivamente 25%, 50% e 75% dos dados da amostra, nesse caso o primeiro quartil demonstra que no minimo 25% dos dados está abaixo dele, segundo quartil 50% está acima e 50% abaixo enquanto o terceiro quartil trás que 75% dos dados estão abaixo dele, os quartis são os dados que se encontram na "caixa" entre o ponto máximo e minimo.

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```
